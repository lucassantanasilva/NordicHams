A flexão Nórdica é um exercício que aumenta a força excêntrica dos músculos isquiotibiais de forma mais eficiente do que outros exercícios aplicados a esse conjunto de músculos, em que é realizado com o voluntário inicialmente ajoelhado, com o tornozelo fixo e tronco na posição vertical. Essa fixação do tornozelo é feita tipicamente acima dos maléolos e pode ser realizada com o auxílio de um profissional ou com algum dispositivo. Posteriormente, o voluntário inicia o movimento de descida com a menor velocidade possível, até que encoste no chão na posição pronada.

Este código irá calcular os valores de torque na articulação do torozelo e comparar com os valores obtidos pelo software Cortex

```python
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import signal
import sys, os
sys.path.insert(1, r'./../functions')
from invdyn2d import invdyn2d
from detecta import detect_onset
```

## Dados obtidos do Cortex e do dinamômetro isocinetico (Biodex)

Neste último item, os dados da coleta obtidos pelo software Cortex e pelo dinamômetro isocinetico serão importados. Estes dados foram registrados ao mesmo tempo em que a coleta foi realizada com a câmera e as marcas refletivas. 

A seguir serão apresentados os resultados mensurados pelo Cortex e pelo Biodex para, em seguida, ser feita uma comparação entre ambos com os resultados obtidos pela dinâmica inversa.

### Cortex

O arquivo gerado pelo cortex está no formato .anc, sendo possível mesmo assim ser aberto como comando `np.loadtxt`. Além disso, os dados foram alocados em variáveis para serem usados posteriormente.

```python
# Lendo o arquivo de dados
filePath = Path('./../data')
data_cortex=np.loadtxt(filePath / 'trialwithoutweight2.anc', skiprows=11)
# Inserindo os dados em suas respectivas variáveis 
name = data_cortex[:,0] # nome
position = data_cortex[:,1] # posição
velocity = data_cortex[:,2] # velocidade
torque = data_cortex[:,3] # torque
sync = data_cortex[:,4]
```

```python
# Data collection parameters
duration= 19.996666
freq_mac    = 300
time_mac= np.linspace(0, duration, torque.shape[0])#Frequência de 200Hz
```

Com o objetivo de comparar o torque de maneira eficaz, este foi convertido para Nm, em que o processo de conversão pode ser vsito na linha de código a seguir.

```python
torque_v = (torque)*(10/65535)
torque_ftlbs = torque_v/0.0098
torque_nm = torque_ftlbs/0.73756 # torque convertido em Nm
```

### Biodex

De forma análoga ao que foi feito com o Cortex, sod dados do Biodex foram importados e os valores de torque armazenados em uma variável, servindo de base para a criação do seu respectivo vetor de tempo.

```python
#Reading
data_biodex=np.loadtxt(r'./../data/RFukuchiISOMRDATA2_2.txt')
```

```python
# Obtendo a variável de torque
torque_biodex = data_biodex[0:-1:3]
time_biodex = np.linspace(0, 30, torque_biodex.shape[0]) #Frequência de 100Hz
```

```python
from detecta import detect_onset
```

```python
threshold = 0
n_above = int(0.1*freq_mac)
n_below = int(0.01*freq_mac)
threshold2 = 10*threshold  # N
n_above2 = int(0.02*freq_mac)
```

```python
ind = detect_onset(torque_nm, threshold, n_above, n_below, threshold2, n_above2, show=True)
```

### Filter signals

```python
# Fit polynomial based on the data
torque_nm2 = torque_nm[ind[0,0]:ind[0,1]]
t = np.linspace(0,torque_nm2.shape[0]/freq_mac,torque_nm2.shape[0])
```

```python
# p = np.polyfit(t, torque_nm2, 9)
# z = np.polyval(p, t)
```

```python
from critic_damp import critic_damp
```

```python
b_cd, a_cd, fc_cd = critic_damp(fcut=5, freq=freq_mac, npass=2, fcorr=True, filt='critic')
torque_nm3 = signal.filtfilt(b_cd, a_cd, torque_nm2)
```

```python
%matplotlib widget
fig, ax = plt.subplots(1, figsize=(10,6))
ax.plot(t, torque_nm2)
ax.plot(t, torque_nm3, 'r')
plt.show()
```

### Calculate max torque

```python
max_torque = 96 # max. torque
idx_max_torque = (np.abs(torque_nm3 - max_torque)).argmin()
```

```python
fig, ax = plt.subplots(1, figsize=(10,6))
ax.plot(time_mac, torque_nm)
ax.plot(time_mac[idx_max_torque+ind[0,0]], torque_nm[idx_max_torque+ind[0,0]], 'k*')
plt.show()
```

Posteriormente, foi feita uma comparação comos valores brutos de torque obtidos por todos os métodos, sem que nenhum tratamento tenha sido empregado nos dados

```python
# Filter Biodex data
b_cd, a_cd, fc_cd = critic_damp(fcut=5, freq=100, npass=2, fcorr=True, filt='critic')
torque_bio2 = signal.filtfilt(b_cd, a_cd, torque_biodex)
```

```python
max_torque_bio = 97.75 # max. torque
idx_max_torque_bio = (np.abs((-torque_bio2/10) - max_torque_bio)).argmin()
```

```python
idx_max_torque_bio
```

```python
fig, ax = plt.subplots(1, figsize=(10,6))
ax.plot(time_biodex,-torque_biodex/10)
ax.plot(time_biodex,-torque_bio2/10,'r')
ax.plot(time_biodex[idx_max_torque_bio],-torque_bio2[idx_max_torque_bio]/10,'k*')
ax.set_xlabel('Time [sec]')
ax.set_ylabel('Torque - Biodex')
ax.grid()
plt.show()
```

### Align two signals based on their correlation

```python
# Filter Cortex signals
b_cd, a_cd, fc_cd = critic_damp(fcut=5, freq=freq_mac, npass=2, fcorr=True, filt='critic')
torque_nm2 = signal.filtfilt(b_cd, a_cd, torque_nm)
```

```python
# Trim Biodex signals to have the same duration than Cortex
torque_bio3 = -torque_bio2/10
torque_bio3 = torque_bio3[time_biodex <= time_mac[-1]]
time_biodex = time_biodex[time_biodex <= time_mac[-1]]
```

```python
# resample Biodex data to the Cortex frequency
freq_cortex = np.round(np.mean(1/np.diff(t)), 0)
torque_bio4 = np.interp(time_mac, time_biodex, torque_bio3)
```

```python
time_biodex2 = np.linspace(0, time_mac[-1], time_mac.shape[0])
```

```python
# Calculate the lag using correlation btw signals
dx = np.mean(np.diff(time_biodex2))
shift = (np.argmax(signal.correlate(torque_bio4, torque_nm2)) - len(torque_nm2)) * dx
```

```python
# Plot Cortex and Biodex signals before and after temporal alignment
fig, ax = plt.subplots(1, figsize=(10,6))
ax.plot(time_mac, torque_nm2,'b', label='Cortex')
ax.plot(time_biodex, torque_bio3,'r', label='Biodex')
ax.plot(time_mac + shift, torque_nm2,'b--', label='Cortex adj')
ax.set_xlabel('Time [sec]')
ax.set_ylabel('Torque [Nm]')
ax.grid()
plt.legend()
plt.show()
```

```python
# Find time interval between thrust and onset of torque signal
inds = detect_onset(torque_bio3, 0.25, n_above=5, n_below=0, show=True)
```

```python
tonset = time_biodex[inds[1,0]]-time_biodex[inds[0,0]]
print('Time lapse between start and onset of torque signal is '+str(tonset.round(2))+' seconds')
```

### Comparação dos resultados

Para que possa ser feita a comparação entre as curvas de torque, é necessário que os dados estejam normalizados. Para tal, foi utilizada a função `detect_onset` que permite acessar os termos do vetor que possuem valores acima de um ponto estipulado. Posteriormente, esse intervalo com valores acima foi inserido em um segundo vetor com 100 termos, representando todo o procedimento de coleta (indo de 0 até 100%).

```python
knee_moms_df = pd.read_csv(filePath / 'knee_moms_kinemat.csv')
knee_moms_df
```

```python
# Plot Cortex and Biodex signals before and after temporal alignment
fig, ax = plt.subplots(1, figsize=(10,6))
fig.suptitle('Comparison of dynamometer torque across methods')
ax.plot(time_mac[time_mac<=6], torque_nm2[time_mac<=6],'b', label='Cortex', lw=3)
ax.plot(knee_moms_df.Time[knee_moms_df.Time<=6], 
        knee_moms_df.TD_ID[knee_moms_df.Time<=6],'r', label='InvDyn')
ax.plot(knee_moms_df.Time[knee_moms_df.Time<=6], 
        knee_moms_df.TS_CMpart[knee_moms_df.Time<=6],'g', label='CMpart')
ax.plot(knee_moms_df.Time[knee_moms_df.Time<=6], 
        knee_moms_df.TS_CMtotal[knee_moms_df.Time<=6],'k', label='CMtotal')
ax.set_xlabel('Time [sec]')
ax.set_ylabel('Torque [Nm]')
ax.grid()
fig.tight_layout(h_pad=.3, w_pad=.05)
plt.legend()
fig.savefig(filePath / 'nordichams_comparison.png', bbox_inches='tight')
plt.show()
```
