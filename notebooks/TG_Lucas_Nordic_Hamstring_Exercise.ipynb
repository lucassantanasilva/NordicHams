## Parâmetros dos segmentos corporais

Um importante conjunto de propriedades mecânicas que auxilia no processo de elaborar um modelo para encontrar o torque no tornozelo é o conjunto dos parâmetros dos segmentos corporais. O corpo humano pode ser analisado como um sistema de corpos conectado, em que cada segmento do corpo possui geometria e parâmetros inerciais distintos. Essas variáveis tais como a massa, centro de massa e momentos de inércia são os chamados parâmetros dos segmentos corporais, sendo uma área da antropometria.

A massa de um corpo é a quantidade de matéria contida nesse corpo, que independe da força da
gravidade agindo neste corpo. Já o momento de inércia é a medida de como a massa de um corpo se distribui em relação a um determinado eixo que passa por um ponto, sendo equivalente a massa em uma rotação. Por fim, o raio de giro representa a distância do eixo a um ponto em que toda massa pode ser concentrada para obter o mesmo momento de inércia do objeto

Com base nisso, foi utilizada a tabela proposta por Zatsiorsky e adaptada por de Leva para obter os parâmetros dos segmentos corporais. Inicialmente após mensurar a massa total dos sujeitos, foram utilizadas as porcentagens referentes a cada segmento. Em outras palavras, como cada parte do corpo representa uma porcentagem da massa total.

### Massa

Com o valor de massa total obtido, foi utilizada a tabela de Zatsiorsky adaptada por de Leva para se obter a massa de cada segmento do corpo do sujeito. Para tal, foi usada como base a equação abaixo:

<span class="notranslate">  
\begin{equation}
m_{i} = M \cdot \text{bsp[i,mass]}
\label{}
\end{equation}
</span>

Em que $m_{i}$ representa a massa do segmento $i$ analisado, $M$ a massa total do sujeito e $bsp[i,mass]$ os valores percentuais tabelados para segmento. A linha a seguir apresenta a aplicação prática desta equação, além de mostrar em sua saída a massa de cada segmento.

<font color='red'>Estas frações de massa também poderiam ser puxadas direto de um arquivo txt que está na pasta data</font>

```python
#BSP male values from Zatsiorsky's model adjusted by de Leva (1996)
m_thigh = M*0.1416
m_trunk = M*0.4346
m_head  = M*0.0694
m_shank = M*0.0433
m_upper_arm = M*0.0271
m_forearm = M*0.0162
m_hand = M*0.0061

print("Thigh:",np.around(m_thigh,3),"Kg")
print("Trunk:",np.around(m_trunk,3),"Kg")
print("Head:",np.around(m_head,3),"Kg")
print("Shank:",np.around(m_shank,3),"Kg")
print("Upper arm:",np.around(m_upper_arm,3),"Kg")
print("Forearm:",np.around(m_forearm,3),"Kg")
print("Hand:",np.around(m_hand,3),"Kg")
```

### Comprimento

O próximo parâmetro calculado foi o comprimento, em que foram obtidos os comprimentos nos eixos x e y e, posteriormente, a resultante dessas duas componentes. Esse procedimento foi feito para os segmentos da perna, coxa e quadril. Para os segmentos da mão, braço, antebraço e cabeça foram utilizados valores tabelados, devido ao fato de terem sido fixados apenas 4 marcas no sujeito.

Para a coxa foi utilizada a seguinte equação com as posições do joelho e quadril, sendo realizada a média desse vetor para encontrar o comprimento do segmento:

<span class="notranslate">
\begin{equation}
Lenght_{Thigh} = \sqrt{lenght_x^2+lenght_y^2} = \sqrt{(Hip_{posX} - Knee_{posX})^2 + (Hip_{posY} - Knee_{posY})}
\label{}
\end{equation}
</span>

Com os dados do quadril e da C7, o mesmo procedimento foi feito para obter o comprimento do tronco:

<span class="notranslate">
\begin{equation}
Lenght_{Trunk} = \sqrt{lenght_x^2+lenght_y^2} = \sqrt{(C7_{posX} - Hip_{posX})^2 + (C7_{posY} - Hip_{posY})}
\label{}
\end{equation}
</span>

Por fim, para os segmentos restantes foram utilizados valores tabelados para o comprimento, representado de forma geral pela equação abaixo:

<span class="notranslate">  
\begin{equation}
Lenght_{i} = H \cdot \text{bsp[i,lenght]}
\label{}
\end{equation}
</span>

Em que $Lenght_{i}$ é o comprimento do segmento desejado, $H$ a altura do sujeito e $bsp[i,lenght]$ o valor tabelado para a altura do segmento $i$.

```python
# Trunk Length
lenght_trunk = np.mean(np.linalg.norm(C7_pos-Hip_pos, axis=1))
print('Length of trunk is ' + str(np.round(lenght_trunk, 2)) + ' meters')
# Thigh Length
lenght_thigh = np.mean(np.linalg.norm(Hip_pos-Knee_pos, axis=1))
print('Length of thigh is ' + str(np.round(lenght_thigh, 2)) + ' meters')
```

### Ângulo dos segmentos

Com o objetivo de garantir que o sujeito realizou a flexão nórdica de forma correta, os ângulos entre cada segmento e o solo foram calculados. De forma ideal, é desejado que os ângulos da coxa e do tronco sejam muito próximos ou mesmo iguais. Para o cálculo dos ângulos foi usada a equação geral a seguir:

<span class="notranslate">    
$$ \theta_{i} = arctan\left(\frac{LenghtY_{i}}{LenghtX_{i}}\right) $$
</span>

Em seguida esses valores foram convertidos de radianos para graus.

```python
theta_thigh = np.arctan2(Hip_posY-Knee_posY,Hip_posX-Knee_posX) * (180 / np.pi)
theta_trunk = np.arctan2(C7_posY-Hip_posY,C7_posX-Hip_posX) * (180 / np.pi)
theta_shank = np.arctan2(Knee_posY-Ankle_posY,Knee_posX-Ankle_posX) * (180 / np.pi)
plt.figure(figsize= [10,5])
plt.plot(t,theta_thigh,label='Coxa')
plt.plot(t,theta_trunk,label='Tronco')
plt.legend(loc='best')
plt.xlabel('Tempo [s]')
plt.ylabel('Ângulo [°]')
plt.title('Ângulo dos segmentos em relação ao solo')
plt.grid()
plt.show()
```

De modo a garantir que realmente a diferença entre os ângulos da coxa e do tronco não apresentam diferença significativa, foi calculada a variação percentual entre esses dois ângulos ao longo do tempo,como pode ser visto na equação a seguir

<span class="notranslate">    
$$ \mathbf{var_{per}} = \left(\frac{\theta_{thigh}}{\theta_{trunk}} - 1\right) \cdot 100 $$
</span>


```python
var_per = ((theta_thigh/theta_trunk) - 1)*100
plt.figure(figsize= [10,5])
plt.plot(t,var_per,label='Variação')
plt.legend(loc='best')
plt.xlabel('Tempo [s]')
plt.ylabel('Variação [%]')
plt.title('Variação percentual entre os ângulos da coxa e do tronco')
plt.grid()
plt.show()
```

### Centro de massa

Outro importante parâmetro a ser estudado neste projeto é o centro de massa, definido como o ponto (posição) onde a soma dos vetores que ligam este ponto a cada partícula, ponderada por sua massa, é zero, em que a massa (a massa inercial) é uma medida quantitativa da resistência de um objeto à aceleração linear. Os valores utilizados para encontrar o centro de massa dos segmentos pode foram obtidos da tabela do modelo de Zatsiorsky adaptado por de Leva em 1996


```python
bsp_Zm = pd.read_csv('./../data/BSPmale_ZdeLeva.txt', index_col=0, sep='\t')
display(bsp_Zm)
```

Tendo isso em vista, como os segmentos da mão, braço e antebraço não possuem marcas, foram utilizados uma série de cálculos para encontrar seus respectivos centros de massa e, posteriormrnte, seus momentos de inércia. O primeiro desses cálculos foi encontrar a altura do pulso, cotovelo e do ombro, para usar como pontos distal e proximal dos segmentos. Por fim, foram utilizados os valores "CM Long" para esses 3 segmentos.

<font color='red'>Acho que aqui, as coordenadas desconhecidas vc precisa usar a articulação distal ou proximal do segmento da marca. Por exemplo, para o cotovelo vc deveria usar o ombro e não o joelho. O ombro deveria usar a marca do quadril, e assim por diante. Tente recalcular assim e veja se melhora</font>

```python
# Altura do ombro em relação ao quadril
shoulder_height = 0.288*H
shoulder_height_y = Hip_posY + shoulder_height*np.sin(theta_trunk*(np.pi/180))
shoulder_height_x = Hip_posX + shoulder_height*np.cos(theta_trunk*(np.pi/180))
shoulder_height_vector = np.array([shoulder_height_x,shoulder_height_y])

# Altura do cotovelo em relação ao ombro
elbow_height = 0.188*H
elbow_height_y = shoulder_height_y - elbow_height*np.sin(theta_trunk*(np.pi/180))
elbow_height_x = shoulder_height_x - elbow_height*np.cos(theta_trunk*(np.pi/180))
elbow_height_vector = np.array([elbow_height_x, elbow_height_y])

# Altura do pulso em relação ao cotovelo
upperarm_height_y = elbow_height_y + lenght_forearm*np.sin(theta_trunk*(np.pi/180))
upperarm_height_x = elbow_height_x + lenght_forearm*np.cos(theta_trunk*(np.pi/180))
upperarm_height_vector = np.array([upperarm_height_x, upperarm_height_y])

# Cálculo do CM Long para o braço, antebraço e mão
forearm_CMLong = 0.4574*lenght_forearm
upperarm_CMLong = 0.5772*lenght_upperarm
hand_CMLong = 0.7900*lenght_hand
```

Posteriormente à isso, as variáveis criadas até agora foram usadas para definir os pontos distal e próximal dos segmentos restantes, sendo eles a perna, coxa, tronco e cabeça.

```python
# Posições diatais e proximais
r_thigh_prox = np.array([Hip_posX,Hip_posY]).T
r_thigh_dis = np.array([Knee_posX,Knee_posY]).T
r_trunk_prox = np.array([Hip_posX,Hip_posY]).T
r_trunk_dis = np.array([C7_posX,C7_posY]).T
r_shank_prox = np.array([Knee_posX,Knee_posY]).T
r_shank_dis =np.array([Knee_posX-lenght_shank,Knee_posY]).T
r_head_prox = np.array([C7_posX,C7_posY]).T
r_head_dis = np.array([C7_posX + lenght_head*np.cos(theta_trunk*(np.pi/180)),C7_posY + lenght_head*np.sin(theta_trunk*(np.pi/180))]).T
```

Com todos os pontos calculados, foi utilizada a equação abaixo para obter o centro de massa de cada segmento:

<span class="notranslate">  
\begin{equation}
rcm_{i} = r_{i,p} + \text{bsp[i,cmp]} \cdot (r_{i,d}-r_{i,p})
\label{}
\end{equation}
</span>

Onde $r_{i,p}$ e $\:r_{i,d}$ são as posições proximal e distal do segmento $i$.

```python
# Centro de massa para os segmentos em x e y
# Antebraço
rcm_forearm_x = elbow_height_x + forearm_CMLong*np.cos(theta_thigh*(np.pi/180))
rcm_forearm_y = elbow_height_y + forearm_CMLong*np.sin(theta_thigh*(np.pi/180))
# Braço
rcm_upperarm_x = shoulder_height_x - upperarm_CMLong*np.cos(theta_thigh*(np.pi/180))
rcm_upperarm_y = shoulder_height_y - upperarm_CMLong*np.sin(theta_thigh*(np.pi/180))
# Mão
rcm_hand_x = upperarm_height_x + hand_CMLong*np.cos(theta_thigh*(np.pi/180))
rcm_hand_y = upperarm_height_y + hand_CMLong*np.sin(theta_thigh*(np.pi/180))

# Centro de massa total
rcm_thigh  = r_thigh_prox + 0.4095*(r_thigh_dis-r_thigh_prox)
rcm_trunk   = r_trunk_prox + 0.4486*(r_trunk_dis-r_trunk_prox)
rcm_shank  = r_shank_prox + 0.4395*(r_shank_dis-r_shank_prox)
rcm_head  = r_head_prox + 0.5976*(r_head_dis-r_head_prox)
rcm_forearm = np.array([rcm_forearm_x, rcm_forearm_y]).T
rcm_upperarm = np.array([rcm_upperarm_x, rcm_upperarm_y]).T
rcm_hand = np.array([rcm_hand_x, rcm_hand_y]).T
rcm_total = ((m_thigh*rcm_thigh) + (m_trunk*rcm_trunk) + (m_forearm*rcm_forearm)+ (m_upper_arm*rcm_upperarm) + (m_hand*rcm_hand)+  (m_head*rcm_head))/(m_upper_arm + m_forearm + m_hand + m_head + m_thigh + m_trunk)
```

### Momento de Inercia

O último parâmetro a ser abordado neste trabalho é o momento de inércia, definido como uma medida quantitativa da resistência à aceleração rotacional em torno de um eixo de uma distribuição de massa no espaço, em que, além dele, precisamos do raio de giro, que é a quantidade que é dada na tabela de parâmetros do segmento corporal, quando falamos de uma fração do comprimento do segmento.

É importante ressaltar que para o cálculo do momento de inércia do tronco foram considerados, além do tronco, ambas asmão, braços e antebraços. De forma análoga, foram considerados ambas as coxas e pernas para seus respectivos momentos de inércia. A seguiinte equação foi utilizada para esse cálculo

<span class="notranslate"> 
\begin{equation}
I_{i,cm} = M \cdot \text{bsp[i,mass]} \cdot \left(\text{bsp[i,rgcm]} \cdot ||r_{i,d}-r_{i,p}||\right)^2
\label{}
\end{equation}
</span>

Em que $M$ é a massa total da pessoa, $||r_{i,d}-r_{i,p}||$ o comprimento do segmento $i$ analisado, $bsp[i,rgcm]$ e $bsp[i,mass]$ os valores tabelados para o raio de giro e a massa de cada segmento, caso o usuário  não tenha calculado os valores de massa anteriormente.

```python
# Raio de giro
Rg_thigh = 0.329
Rg_trunk = 0.372
Rg_head = 0.362
Rg_shank = 0.251

# Momento de inércia
Icm_thigh = (m_thigh*2)*((Rg_thigh*(np.mean(np.linalg.norm(r_thigh_dis-r_thigh_prox, axis=1)))**2)) #Sag
Icm_trunk = (m_trunk+m_forearm*2+m_hand*2+m_upper_arm*2)*((Rg_trunk*(np.mean(np.linalg.norm(r_trunk_dis-r_trunk_prox, axis=1)))**2)) #Sag 
Icm_shank = (m_shank*2)*((Rg_shank*(np.mean(np.linalg.norm(r_shank_dis-r_shank_prox, axis=1)))**2))
Icm_head = (m_head)*((Rg_head*(np.mean(np.linalg.norm(r_head_dis-r_head_prox, axis=1)))**2))

# Resultados dos cálculos
print ('\033[1m''Momentos de Inércia\n''\033[0m')
print('Thigh:',np.around(Icm_thigh,3),"kgm2")
print('Trunk:',np.around(Icm_trunk,3),"kgm2")
print('Shank:',np.around(Icm_shank,3), "kgm2")
print("Head:",np.around(Icm_head, 3),"kgm2")
```

## Dinâmica Inversa

A partir da posição e orientação dos segmentos do corpo, da medição das forças externas é possível estimar as forças e torques internos responsáveis por causar os movimentos analisados, por meio da abordagem da dinâmica inversa. Esse método combina as medidas cinemáticas, obtidas por um sistema de captura de movimento, com as forças externas ao movimento, obtidas por meio de plataformas de força e parâmetros inerciais dos segmentos, para estimar as forças e torques nas articulações, sendo usado em modelos com mais de um segmento. 

Para simplificar o problema, os segmentos do corpo humano são considerados corpos rígidos cuja deformação é desconsiderada, o que implica que as distâncias entre todos os pares de pontos nesse corpo,além de sua massa total, são consideradas constantes. 

Como resolver as equações para encontrar as forças e torques nas articulações de forma manual é um processo demorado e pode levar à erros, foi utilizada a função `invdyn2d.py`, que fornece a força e torque do segmento desejado, em que foi aplicada três vezes, oara a cabeça, tronco e coxa.

```python
# Cabeça
Fp_head, Mp_head = invdyn2d(rcm_head, r_head_dis, r_head_prox, acm_trunk, atheta_trunk, m_head, Icm_head, np.array([0,0]), 0)
# Tronco
Fp_trunk, Mp_trunk = invdyn2d(rcm_trunk, r_trunk_dis, r_trunk_prox, acm_trunk, atheta_trunk, m_trunk, Icm_trunk,  -Fp_head, -Mp_head)
# Coxa
Fp_thigh, Mp_thigh = invdyn2d(rcm_thigh, r_thigh_prox, r_thigh_dis, acm_thigh, atheta_thigh, m_thigh, Icm_thigh, -Fp_trunk, -Mp_trunk)
```

## Modelos estáticos

Além da dinâmica inversa, foram elaborados modelos mais simples para estimar o torque, em que após os cálculos serão feitas comparações entre todos os modelos, de modo a verificar a eficácia desses modelos mais simples. 

### Modelo estático completo

Esse primeiro modelo simplificado utiliza a mesma equação de torque geral anterior, em que a distância será do joelho até o centro de massa total da pessoa. Esse centro de massa considera ambos os braços, antebraços, mãos e coxas para seu cálculo, em que é calculado usando a equação abaixo:

<span class="notranslate">    
$$ \vec{r}_{cm} = \frac{\sum\limits_{i=1}^nm_i{\vec{r}_i}}{m_T} $$
</span>

Em que $\vec{r}_{cm}$ representa o centro de massa total, $\sum\limits_{i=1}^nm_i{\vec{r}_i}$ o produto da massa pelo vetor posiçãode cada segmento pelo seu respectivo centro de massa e $m_T$ a massa total. A obtenção do torque foi feita com base no seguinte Diagrama de Corpo Livre.

```python
# Massa total dos segmentos considerados
MTotal = (m_thigh*2) + m_trunk + m_head + (m_forearm*2) + (m_hand*2) + (m_upper_arm*2)

# Centro de massa do modelo estático
rcm_total_static = (((m_thigh*2)*rcm_thigh) + ((m_trunk+m_forearm*2+m_hand*2+m_upper_arm*2)*rcm_trunk) + (m_head*rcm_head))/MTotal

# Distância do joelho até o centro de massa total
L_rcm_Total = rcm_total_static - r_thigh_dis
L_rcm = np.sqrt(L_rcm_Total[:,0]*L_rcm_Total[:,0] + L_rcm_Total[:,1]*L_rcm_Total[:,1])

# Cálculo do torque (VERIFICAR ESSES VALORES ABAIXO DE FORÇA E TORQUE DEPOIS NO LAB)
# M = MTotal*g*L_rcm*np.cos(theta_thigh*(np.pi/180))
M = MTotal*g*L_rcm*np.cos(theta_thigh) - (m_shank*2)*g*(lenght_shank/2)*np.cos(theta_shank)
F = M/0.46
Mt = F*0.308
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10,6))

ax.plot(t[0:len(Mt)],Mt,label = 'Moment')
ax.grid()
plt.xlabel('time (s)')
plt.ylabel(r'Moment (Nm)')
ax.legend(loc='best')
plt.suptitle('Modelo estático completo', fontsize=30);
plt.show()
```

### Modelo estático simplificado

Esse segundo modelo é ainda mais simples, pois não considera o centro de massa total. Nele a pessoa é considerada uma barrra, em que seu comprimento éa soma dos comprimentos da coxa, tronco e cabeça. Como o centro de massa de uma barra fica na metade de seu comprimento, a distância seria simplesmente a metade do comprimento dessa barra, em que o torque calculado seria dado por:

<span class="notranslate">    
$$ M = m_T \cdot g \cdot \left(\frac{L}{2}\right)  \cdot cos(\theta_{thigh}) - m_{shank}\cdot g \cdot \left(\frac{L_{shank}}{2}\right)\cdot cos(\theta_{shank})$$
</span>

Em que $g$ é a aceleração da gravidade, $L$ o comprimento total da barra calculado e $\theta$ é o ângulo da coxa. O cálculo do torque seguiu o mesmo DCL adotado anteriormente.

```python
# Cálculo do comprimento total da barra
L = lenght_head + lenght_thigh + lenght_trunk

# Cálculo do torque (VERIFICAR ESSES VALORES ABAIXO DE FORÇA E TORQUE DEPOIS NO LAB)
#M2 = MTotal*g*(L/2)*np.cos(theta_thigh*(np.pi/180))
M2 = MTotal*g*(L/2)*np.cos(theta_thigh) - (m_shank*2)*g*(lenght_shank/2)*np.cos(theta_shank)
F2 = M2/0.46
Mt2 = F2*0.308
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10,6))

ax.plot(t,Mt2, label = 'Moment')
ax.grid()
plt.xlabel('time (s)')
plt.ylabel(r'Moment (Nm)')
ax.legend(loc='best')
plt.suptitle('Modelo estático simplificado', fontsize=30);
plt.show()
```

### Comparação dos resultados

Para que possa ser feita a comparação entre as curvas de torque, é necessário que os dados estejam normalizados. Para tal, foi utilizada a função `detect_onset` que permite acessar os termos do vetor que possuem valores acima de um ponto estipulado. Posteriormente, esse intervalo com valores acima foi inserido em um segundo vetor com 100 termos, representando todo o procedimento de coleta (indo de 0 até 100%).

```python
# Detect events

# Cortex
inds = detect_onset(torque_nm, 10, n_above=10, n_below=0, show=False)
xx = torque_nm[inds[0][0]:inds[0][1]]
t  = np.linspace(0, 100, len(xx))  # time vector for the original data
tn = np.linspace(0, 100, 101)
yn = np.empty(shape=(101))
yn[:] = np.interp(tn, t, xx)

# Biodex
torque_bio = -torque_biodex/10
inds = detect_onset(torque_bio, 10, n_above=10, n_below=0, show=False)
xx = torque_bio[inds[0][0]:inds[0][1]]
t  = np.linspace(0, 100, len(xx))  # time vector for the original data
tn2 = np.linspace(0, 100, 101)
yn2 = np.empty(shape=(101))
yn2[:] = np.interp(tn2, t, xx)
```
